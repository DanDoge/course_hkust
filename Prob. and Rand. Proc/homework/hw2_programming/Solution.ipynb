```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('iris data.xlsx')
data
```

```python
fig, ax = plt.subplots()
n, bins, patches = ax.hist(data['Petal length'], bins=np.arange(0, 6, 0.2), density=True)
```

```python
mean = data.mean()['Petal length']
```

```python
std = data.std()['Petal length']
mean, std
```

```python
y = ((1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-0.5 * (1 / std**2 * (bins - mean))**2))
```

```python
ax.plot(bins, y, '--')
fig
```
